```python
#En principio buscamos conocer el comportamiento de la totalidad de la muestra, en relación a cada ítem.
#Primero, se hará un análisis de la variable "Tiempo".

#Creamos dos variables. Data --> lista vacía para que los datos correspondientes al tiempo se almacenen allí.
#La variables item tendrá el valor 1, ya que corresponde al primer item del Robotest.

data = []
item = 1

#Se crea un for loop para que se recorra cada item del Robotest. Este loop irá calculando al media de las variables tiempo.
# y lo redondeará en dos decimales. El resultado será añadido a la variable "data".
# Este loop se repetirá hasta que item sea 40, si es 41 se rompe.

for x in df[str(item) + ' Tiempo']:
        mediana = round(df[str(item) + ' Tiempo'].median(), 2)
        data.append(mediana)
        item += 1
        if item == 41:
            break

#Se crea un dicccionario en la variable d, donde se almacena la llave "Medias - Tiempo" con los datos almacenados en data.
#En la variable trail se almacena un rango de números desde 1 hasta 40 (41 está excluido)
#En la variable data se almacena un data prame, con el diccionario almacenado en d.

d = {'Medianas - Tiempo': data}
trial = range(1,41)
data = pd.DataFrame(data = d)


#Con todos estos datos creamos una figura.
#Primero se crea una figura vacía donde se especifica el tamaño de la misma. 
#A partir de eso se crea un gráfico de línea que muestra el tiempo promedio de respuesta. 

plt.figure(figsize= (14, 5))
sns.lineplot(data = data,
             x = trial,
             y = 'Medianas - Tiempo',
             marker = 'o',
             markersize= 8,
             color = 'seagreen'
             )
plt.xticks(range(1, 41))
plt.ylim(ymin=0)
plt.title('Robotest - Tiempo de respuesta (mediana) - Muestra Total', fontsize = 10, fontweight = 'bold')
plt.xlabel('Item', fontsize = 10, labelpad = 12)
plt.ylabel('Tiempo (ms)', fontsize = 10, labelpad = 12)

plt.show()
```

```python
#Análisis por grado.
#Desarrollo de una función que permite hacer un análisis de las variables "Puntaje" y "Tiempo" de cada Trial del Robotest.

#La función se llamará "grado" y entre paréntesis se especifican los parámetros que necesita para su funcionamiento.
#Para que la función funcione correctamente, será necesario que se importe la libreria Pandas, Numpy, Matplotlib.pyplot 
# y Seaborn.


def grado(data = None, escuela = str, grado = int, titl = False, 
          stitl_1 = False, stitl_2 = False, trial = int, g_pje = True):
    
    """Esta función permite realizar análisis descriptivo de las variables 'Puntaje' y 'Tiempo'.
    Para el funcionamiento de la misma, se necesitarán los siguientes parámetros:
        data = None. Dataframe de donde se sacarán los datos.
        escuela = str. Identificación de la escuela, ej. 'E1', 'E2'.
        grado = int. Número que representa el grado a analizar. 
        titl = Bool. Título del análisis. Por default es 'False'.
        stitl_1 = Bool. Subtítulo de análisis de Puntaje. Por default es 'False'.
        stitl_2 = Bool. Subtítulo de análisis de Tiempo. Por default es 'False'.
        trial = int. Número que representa el trial a analizar.
        g_pje = Bool. Gráfico que muestra la cantidad de niños que respondió correcta e incorrectamente el trial. 
                Por default es "True".
       
    La función devuelve:
        El número de niños que respondió correcta e incorrectamente el trial. El mismo puede estar acompañado con
        un gráfico en caso de que el usuario lo requiera.
        Análisis descriptivo de la variable 'Tiempo', junto con un histograma para observar distribución"""

#En la variable "e" se guarda el recorte del df de la escuela y curso
   
    e = data[(data['Escuelas'] == escuela) & (df['Curso'] == grado)]
    
    
#Se presenta la posibilidad de asignar un título al análisis. El mismo permitirá conocer el trial que se esté analizando
#del Test "Robotests".

    if titl == True:
        print((' '*49) + 'Análisis del Trial N°'+ str(trial) + ' - Robotest' + ('\n')*2)
    
    print('Escuela: ' + escuela + '.')
    print('Grado: ' + str(grado) +'.' + ('\n')*2)
   
    puntaje = (str(trial) + ' Puntaje')
    
    if stitl_1 == True:
        print('Puntaje:' + ('\n'))

    positivos = 0
    negativos = 0
    for x in e[puntaje]:
        if x == 1:
            positivos += 1
        if x == 0:
            negativos +=1

    if positivos == 0:
        resultado1= 'Ningún niño ha respondido correctamente.'
    elif positivos == 1:
        resultado1 = 'Sólo un niño ha respondido correctamente.'
    else:
        resultado1 = (str(positivos) + ' niños han respondido correctamente.') 
    

    if negativos == 0:
        resultado2 = 'Ningún niño ha respondido incorrectamente.'
    elif negativos == 1:
        resultado2 = 'Sólo un niño ha respondido incorrectamente.'
    else:
        resultado2 = (str(negativos) + ' niños han respondido correctamente.')
    
    print(resultado1 + '\n' + resultado2)

#Gráfico que nos muestra la cantidad de niños que han respondido correcta e incorrectamente.
  
    if g_pje == True:  
        
        sns.histplot(data = e, 
                     x = puntaje,
                     hue = puntaje,
                     palette = 'Set2',
                     shrink= 5)

        if negativos == 0:
            xticks = ([1])
        elif positivos == 0:
            xticks = ([0])
        else:
            xticks = ([0, 1])
        
        plt.xticks(xticks)
        plt.yticks([0, 5, 10, 15, 20])
        plt.title('Escuela: ' + escuela + ' - Robotest')
        plt.xlabel('Puntaje', fontsize = 10, labelpad = 12)
        plt.ylabel('Niños', fontsize = 10, labelpad = 12)
        plt.legend(loc='upper left', labels=['Correcto', 'Incorrecto'])
        plt.show()
    else: 
        None
    
#Pasamos a realizar un análisis descriptivo de la variable "Tiempo".
    if stitl_2 == True:
        print('\n' + 'Tiempo:' + '\n')

    tiempo = (str(trial) + ' Tiempo')
    print(str(round(e[tiempo].describe(),2)))


#Realizamos un gráfico para visualizar el comportamiento del tiempo

    sns.histplot(data = e,
                 x = tiempo,
                 color = 'seagreen',
                 stat = 'count')
    plt.xlabel('Tiempo', fontsize = 10, labelpad = 12)
    plt.ylabel('Niños', fontsize = 10, labelpad = 12)
    plt.title('Escuela: '+ escuela + ' - Robotest')
    plt.show()

    from matplotlib import rcParams
    rcParams['figure.figsize'] = 5, 5
```

```python
#Análisis por edad
#Desarrollo de una función que permite hacer un análisis de las variables "Puntaje" y "Tiempo".

def edad(data = None, escuela = str, grado = int, edad = int, titl = False, 
         stitl_1 = False, stitl_2 = False, trial = int, g_pje = True):
 
    """Esta función permite realizar análisis descriptivo de las variables 'Puntaje' y 'Tiempo', según la edad.
    Para el funcionamiento de la misma, se necesitarán los siguientes parámetros:
        data = None. Dataframe de donde se sacarán los datos.
        escuela = str. Identificación de la escuela, ej. 'E1', 'E2'.
        grado = int. Número que representa el grado a analizar. 
        edad = int. Número que representa la edad redondeada de los niños.
        titl = Bool. Título del análisis. Por default es 'False'.
        stitl_1 = Bool. Subtítulo de análisis de Puntaje. Por default es 'False'.
        stitl_2 = Bool. Subtítulo de análisis de Tiempo. Por default es 'False'.
        trial = int. Número que representa el trial a analizar. 
        g_pje = Bool. Gráfico que muestra la cantidad de niños que respondió correcta e incorrectamente el trial. 
                Por default es "True"
       
    La función devuelve:
        El número de niños que respondió correcta e incorrectamente el trial, junto con un histograma. 
        Análisis descriptivo de la variable 'Tiempo', junto con un histograma para observar distribución"""

    curso = df[(df['Escuelas'] == escuela) & (df['Edad en años redondeada'] == edad)]

    
    if titl == True:
               
        print((' '*49) + 'Análisis del Trial N°'+ str(trial) + ' - Robotest' + ('\n')*2)
    
    print('Escuela: ' + escuela + '.')
    print('Grado: ' + str(grado) +'.')
    print('Edad: ' + str(edad) + '.' + ('\n')*2)
   
    puntaje = (str(trial) + ' Puntaje')
    if stitl_1 == True:
        print('Puntaje:' + ('\n'))

    positivos = 0
    negativos = 0
    for x in curso[puntaje]:
        if x == 1:
            positivos += 1
        if x == 0:
            negativos +=1

    if positivos == 0:
        resultado1= 'Ningún niño ha respondido correctamente.'
    elif positivos == 1:
        resultado1 = 'Sólo un niño ha respondido correctamente.'
    else:
        resultado1 = (str(positivos) + ' niños han respondido correctamente.') 
    

    if negativos == 0:
        resultado2 = 'Ningún niño ha respondido incorrectamente.'
    elif negativos == 1:
        resultado2 = 'Sólo un niño ha respondido incorrectamente.'
    else:
        resultado2 = (str(negativos) + ' niños han respondido correctamente.')
    
    print(resultado1 + '\n' + resultado2)

#Gráfico que nos muestra la cantidad de niños que han respondido correcta e incorrectamente.
  
    if g_pje == True: 
        
        sns.histplot(data = curso, 
                     x = puntaje,
                     hue = puntaje,
                     palette = 'Set2',
                     shrink= 5)

    
        if negativos == 0:
            xticks = ([1])
        elif positivos == 0:
            xticks = ([0])
        else:
            xticks = ([0, 1])
        
        plt.xticks(xticks)
        plt.yticks([0, 5, 10, 15, 20])
        plt.title('Escuela: ' + escuela + ' - Robotest')
        plt.xlabel('Puntaje', fontsize = 10, labelpad = 12)
        plt.ylabel('Niños', fontsize = 10, labelpad = 12)
        plt.legend(loc='upper left', labels=['Correcto', 'Incorrecto'])
        plt.show()
    
    else:
        None

    
#Pasamos a realizar un análisis descriptivo de la variable "Tiempo".
    if stitl_2 == True:
        print('\n' + 'Tiempo:' + '\n')

    tiempo = (str(trial) + ' Tiempo')
    print(str(round(curso[tiempo].describe(),2)))


#Realizamos un gráfico para visualizar el comportamiento del tiempo

    sns.histplot(data = curso,
                 x = tiempo,
                 color = 'seagreen',
                 stat = 'count')
    plt.xlabel('Tiempo', fontsize = 10, labelpad = 12)
    plt.ylabel('Niños', fontsize = 10, labelpad = 12)
    plt.title('Escuela: '+ escuela + ' - Robotest')
    plt.show()
```
